```python
import pandas as pd
import requests
from PIL import Image
import io
import numpy as np
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
from tensorflow.keras.preprocessing import image
from sklearn.metrics.pairwise import cosine_similarity
import os
import matplotlib.pyplot as plt
```

```python
class RecallDataProcessor:
    def __init__(self, csv_path):
        """Initialize with path to CSV file"""
        self.df = pd.read_csv(csv_path)
        print(f"Loaded dataset with {len(self.df)} records")
        self.model = ResNet50(weights='imagenet', include_top=False, pooling='avg')
        self.image_features = {}

    def is_valid_image_url(self, url):
        """Check if URL contains valid image extension"""
        if not isinstance(url, str):
            return False
        valid_extensions = ['.jpg', '.jpeg', '.png']
        return any(ext in url.lower() for ext in valid_extensions)

    def display_image_from_url(self, url):
        """Display image from URL"""
        try:
            response = requests.get(url)
            img = Image.open(io.BytesIO(response.content))
            plt.imshow(img)
            plt.axis('off')
            plt.show()
        except Exception as e:
            print(f"Error displaying image: {str(e)}")

    def download_and_process_image(self, url):
        """Download image from URL and convert to feature vector"""
        try:
            response = requests.get(url, timeout=10)
            img = Image.open(io.BytesIO(response.content))
            img = img.convert('RGB')  # Convert to RGB if needed
            img = img.resize((224, 224))  # ResNet50 expected size
            img_array = image.img_to_array(img)
            img_array = np.expand_dims(img_array, axis=0)
            img_array = preprocess_input(img_array)
            features = self.model.predict(img_array)
            return features.flatten()
        except Exception as e:
            print(f"Error processing image from {url}: {str(e)}")
            return None

    def process_dataset(self, limit=None):
        """Process all valid images in dataset and store their features"""
        print("Starting dataset processing...")
        valid_count = 0

        # If limit is set, only process that many records
        df_to_process = self.df[:limit] if limit else self.df

        for idx, row in df_to_process.iterrows():
            image_url = row['Image URL']
            if self.is_valid_image_url(image_url):
                print(f"Processing image {idx + 1}: {image_url}")
                features = self.download_and_process_image(image_url)
                if features is not None:
                    self.image_features[idx] = features
                    valid_count += 1

        print(f"Processed {valid_count} valid images out of {len(df_to_process)} records")
        return self.image_features

    def find_similar_images(self, query_image_path, top_k=5):
        """Find similar images from the dataset given a query image"""
        try:
            # Process query image
            query_img = Image.open(query_image_path)
            query_img = query_img.convert('RGB')
            query_img = query_img.resize((224, 224))
            query_array = image.img_to_array(query_img)
            query_array = np.expand_dims(query_array, axis=0)
            query_array = preprocess_input(query_array)
            query_features = self.model.predict(query_array).flatten()

            # Calculate similarities
            similarities = {}
            for idx, features in self.image_features.items():
                similarity = cosine_similarity(
                    query_features.reshape(1, -1),
                    features.reshape(1, -1)
                )[0][0]
                similarities[idx] = similarity

            # Sort by similarity and get top_k matches
            sorted_matches = sorted(similarities.items(), key=lambda x: x[1], reverse=True)[:top_k]

            # Return results with confidence scores and display images
            results = []
            for idx, confidence in sorted_matches:
                record = self.df.iloc[idx].to_dict()
                print(f"\nMatch {len(results) + 1} (Confidence: {confidence:.2f})")
                print(f"Product: {record['Name']}")
                print(f"Hazard: {record['Hazard']}")
                print(f"Reason for recall: {record['Reason for recall']}")
                self.display_image_from_url(record['Image URL'])

                results.append({
                    'confidence': float(confidence),
                    'recall_info': {
                        'name': record['Name'],
                        'supplier': record['Supplier Name'],
                        'published_date': record['Published Date'],
                        'hazard': record['Hazard'],
                        'reason_for_recall': record['Reason for recall'],
                        'consumer_should_do': record['Consumer should do'],
                        'image_url': record['Image URL']
                    }
                })

            return results
        except Exception as e:
            print(f"Error processing query image: {str(e)}")
            return []

# Example usage
# Replace with your CSV path from Google Drive
CSV_PATH = '/content/Trial Task - AU Product Safety Recalls - 2020 - 2024 - Processed.csv'

```

```python
processor = RecallDataProcessor(CSV_PATH)

# Process first 100 images (for testing)
processor.process_dataset(limit=100)
```
